## evaluation

```python
fig = plt.figure()

plt.title("Evaluation")

plt.xlabel("iteration")
plt.ylabel("ratio")

x = np.arange(20)
losses = np.ones(20) - wins - draws
# no_loss = np.zeros(50) + wins +draws

plt.plot(x, wins, label="win ratio")
plt.plot(x, draws, label="draw ratio")
#plt.plot(x, no_loss, label="no loss ratio")
plt.plot(x, losses, label="loss ratio")

plt.legend(loc=2)
plt.show()
```

```python
import numpy as np
```

```python
dirichlet_sample_02 = np.random.dirichlet([0.2] * 9)
dirichlet_sample_04 = np.random.dirichlet([0.4] * 9)
dirichlet_sample_06 = np.random.dirichlet([0.6] * 9)
dirichlet_sample_08 = np.random.dirichlet([0.8] * 9)
dirichlet_sample_10 = np.random.dirichlet([1.0] * 9)
dirichlet_sample_12 = np.random.dirichlet([1.2] * 9)
dirichlet_sample_14 = np.random.dirichlet([1.4] * 9)
dirichlet_sample_16 = np.random.dirichlet([1.6] * 9)
```

```python
print("alpha = 0.2: {}" .format(dirichlet_sample_02))
print("alpha = 0.4: {}" .format(dirichlet_sample_04))            
print("alpha = 0.6: {}" .format(dirichlet_sample_06))            
print("alpha = 0.8: {}" .format(dirichlet_sample_08))            
print("alpha = 1.0: {}" .format(dirichlet_sample_10))            
print("alpha = 1.2: {}" .format(dirichlet_sample_12))            
print("alpha = 1.4: {}" .format(dirichlet_sample_14))
print("alpha = 1.6: {}" .format(dirichlet_sample_16))            
```

```python
fig = plt.figure()

plt.title("Dirichlet")

plt.xlabel("move")
plt.ylabel("probability")

x = np.arange(9)

plt.plot(x, dirichlet_sample_02, label="alpha = 0.2")
#plt.plot(x, dirichlet_sample_04, label="alpha = 0.4")
#plt.plot(x, dirichlet_sample_06, label="alpha = 0.6")
#plt.plot(x, dirichlet_sample_08, label="alpha = 0.8")
plt.plot(x, dirichlet_sample_10, label="alpha = 1.0")
#plt.plot(x, dirichlet_sample_12, label="alpha = 1.2")
#plt.plot(x, dirichlet_sample_14, label="alpha = 1.4")
plt.plot(x, dirichlet_sample_16, label="alpha = 1.6")

plt.legend()
```
