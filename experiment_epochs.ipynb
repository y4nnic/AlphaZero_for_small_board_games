### Training loss

```python
# import data
train_loss_1 = pd.read_csv("tictactoe_epochs/run_tictactoe_1_epoch-tag-loss.csv")
train_loss_1 = train_loss_1['Value'].values

train_loss_2 = pd.read_csv("tictactoe_epochs/run_tictactoe_3_epochs-tag-loss.csv")
train_loss_2 = train_loss_2['Value'].values

train_loss_3 = pd.read_csv("tictactoe_epochs/run_tictactoe_5_epochs-tag-loss.csv")
train_loss_3 = train_loss_3['Value'].values

train_loss_4 = pd.read_csv("tictactoe_epochs/run_tictactoe_10_epochs-tag-loss.csv")
train_loss_4 = train_loss_4['Value'].values

# length 50 
print(train_loss_1.shape)
print(train_loss_2.shape)
print(train_loss_3.shape)
print(train_loss_4.shape)

print("short:")
print(train_loss_1.shape)
train_loss_2_short = (train_loss_2[0::3] + train_loss_2[1::3] + train_loss_2[2::3])/3
print(train_loss_2_short.shape)

train_loss_3_short = (train_loss_3[0::3] + train_loss_3[1::3] + train_loss_3[2::3])/3
print(train_loss_3_short.shape)


train_loss_4_short = (train_loss_4[0::10] + 
                      train_loss_4[1::10] + 
                      train_loss_4[2::10] + 
                      train_loss_4[3::10] + 
                      train_loss_4[4::10] + 
                      train_loss_4[5::10] + 
                      train_loss_4[6::10] + 
                      train_loss_4[7::10] + 
                      train_loss_4[8::10] + 
                      train_loss_4[9::10])/10
print(train_loss_4_short.shape)
```
